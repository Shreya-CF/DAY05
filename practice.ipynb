```python
#Linear Search Function
def linear_search(numbers, target):

    for number in numbers:
        if number == target:
            return True

    return False


numbers = [10, 20, 30, 40, 50]

print(linear_search(numbers, 40))
print(linear_search(numbers, 100))
```

```python
#searching algorithms
# Function definition
def linear_search(numbers, target):
    for number in numbers:
        if number == target:
            return True
    return False


def binary_search(numbers, target):
    left = 0
    right = len(numbers) - 1

    while left <= right:
        mid = (left + right) // 2

        if numbers[mid] == target:
            return mid
        elif numbers[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1


# Main Program
numbers = [5, 10, 15, 20, 25]

print("Linear Search for 20:", linear_search(numbers, 20))
print("Linear Search for 100:", linear_search(numbers, 100))

sorted_numbers = [10, 20, 30, 40, 50, 60]

print("Binary Search for 40:", binary_search(sorted_numbers, 40))
print("Binary Search for 90:", binary_search(sorted_numbers, 90))
```

```python
from collections import deque
# Queue - FIFO


queue = deque()

# Enqueue
queue.append(10)
queue.append(20)
queue.append(30)

print("After Enqueue:", queue)

# Peek
print("Front Element:", queue[0])

# Dequeue
removed = queue.popleft()

print("Removed:", removed)
print("After Dequeue:", queue)

# Is Empty
print("Is Empty:", len(queue) == 0)
```

```python
# Bubble Sort
numbers = [5, 3, 8, 4]

n = len(numbers)

for i in range(n):

    for j in range(n - i - 1):

        if numbers[j] > numbers[j + 1]:

            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]

print("Sorted List:", numbers)
```

```python
#Optimized Bubble Sort
def bubble_sort(numbers):

    n = len(numbers)

    for i in range(n):

        swapped = False

        for j in range(n - i - 1):

            if numbers[j] > numbers[j + 1]:

                numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]

                swapped = True

        if not swapped:
            break

    return numbers


numbers = [1, 2, 3, 4, 5]

print(bubble_sort(numbers))
```

```python
# Selection Sort
numbers = [64, 25, 12, 22, 11]

n = len(numbers)

for i in range(n):

    min_index = i

    for j in range(i + 1, n):

        if numbers[j] < numbers[min_index]:
            min_index = j

    numbers[i], numbers[min_index] = numbers[min_index], numbers[i]

print("Sorted List:", numbers)
```

```python
# Insertion Sort
numbers = [5, 2, 4, 6, 1, 3]

n = len(numbers)

for i in range(1, n):

    key = numbers[i]

    j = i - 1

    while j >= 0 and numbers[j] > key:

        numbers[j + 1] = numbers[j]

        j -= 1

    numbers[j + 1] = key

print("Sorted List:", numbers)
```

```python
# Merge Sort
def merge_sort(arr):

    # Base Case: If the list has 0 or 1 element, it's already sorted
    if len(arr) <= 1:
        return arr

    # Find the middle index
    mid = len(arr) // 2

    # Divide the array into two halves
    left = arr[:mid]
    right = arr[mid:]

    # Recursively sort both halves
    left = merge_sort(left)
    right = merge_sort(right)

    # Merge the sorted halves
    return merge(left, right)


def merge(left, right):

    result = []

    i = 0
    j = 0

    # Compare elements from both lists
    while i < len(left) and j < len(right):

        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # Add remaining elements from left list
    while i < len(left):
        result.append(left[i])
        i += 1

    # Add remaining elements from right list
    while j < len(right):
        result.append(right[j])
        j += 1

    return result


# -------------------------------
# Main Program
# -------------------------------

numbers = [38, 27, 43, 3, 9, 82, 10]

print("Original List:", numbers)

sorted_numbers = merge_sort(numbers)

print("Sorted List:", sorted_numbers)
```

```python
#Recursive Fibonacci
def fibonacci(n):

    if n <= 1:
        return n

    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(6))
```
